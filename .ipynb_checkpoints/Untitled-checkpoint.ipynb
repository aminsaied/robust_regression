# Experiment

```python
def simulate_data(N, intercept, slope, nu, sigma2):
    """Simulate linear data..."""
    np.random.seed()
    
    # Create a pandas DataFrame with column 'x' containing
    # N uniformly sampled values between 0.0 and 1.0
    interval = np.linspace(0,1, num=2*N)
    sample = np.random.choice(interval, size=N, replace=False)
    df = pd.DataFrame({"x": sample})

    # generate y values using linear model
    linear_map = lambda x: intercept + slope*x
    df['y'] = linear_map(df['x']) + sigma2*np.random.standard_t(nu, N)
    
    return df
```

```python
nu = 3
a = 10*np.random.randn()
b = 10*np.random.rand()
df = simulate_data(100, b, a, nu, 2)
X = sm.add_constant(np.array(df['x']))
y = np.array(df['y']).reshape(-1,1)

true = np.array([[b],[a]])

model = ECME(y, X, compare=True)
model.fit(30)

model_s = ECME(y, X, use_sigma2=True)
model_s.fit(30)


print(true[0])
print(model.B[0])
print(model_s.B[0])
print(model.B_huber_1[0])
print(model.B_huber_5[0])
print()
print(true[1])
print(model.B[1])
print(model_s.B[1])
print(model.B_huber_1[1])
print(model.B_huber_5[1])


# B_ols = model.B_ols
# B_hubert_1 = model.B_huber_1
# B_hubert_5 = model.B_huber_5
# B_ecme = model.B
```

# Simulation

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

from em_algorithm import ECME
from utils import from_dataframe

import matplotlib.pyplot as plt
%matplotlib inline

import simulation
```

```python
df = simulation.simulate_data(100, 0, 1, 6, 3)
y, X = from_dataframe(df)
model = ECME(y, X, use_sigma2=True)
model.fit()
print(model.B)
print(model.sigma2)
print(model.nu)
```

```python
# to record errors
models = ['ecme', 'huber1', 'huber2', 'huber3', 'huber4', 'ols']
errors = { model : {'a':[], 'b':[]} for model in models}
# errors['ecme'] = {'a':[], 'b':[]}
# errors['ols'] = {'a':[], 'b':[]}
# errors['huber1'] = {'a':[], 'b':[]}
# errors['huber2'] = {'a':[], 'b':[]}
# errors['huber3'] = {'a':[], 'b':[]}
# errors['huber4'] = {'a':[], 'b':[]}

# dof estimate
dof_estimates = []

nus = np.linspace(0.5, 10, 10)
for nu in nus:
    
    nu_errors = { model : {'a':[], 'b':[]} for model in models}    
    dof_estimates_tmp = []
    
    for _ in range(20):
        a = 10*np.random.randn()
        b = 10*np.random.rand()
        sigma2 = 2*np.random.rand()
        df = simulation.simulate_data(100, b, a, nu, sigma2)
        
        y, X = from_dataframe(df)
        
        model = ECME(y, X, compare=True, use_sigma2=True)
        model.fit()
        
        # intercept
        nu_errors['ecme']['b'].append(np.abs((model.B[0] - b)/b))
        nu_errors['ols']['b'].append(np.abs((model.B_ols[0] - b)/b))
        nu_errors['huber1']['b'].append(np.abs((model.B_huber_1[0] - b)/b))
        nu_errors['huber2']['b'].append(np.abs((model.B_huber_2[0] - b)/b))
        nu_errors['huber3']['b'].append(np.abs((model.B_huber_3[0] - b)/b))
        nu_errors['huber4']['b'].append(np.abs((model.B_huber_4[0] - b)/b))
    
        # slope
        nu_errors['ecme']['a'].append(abs((model.B[1] - a)/a))
        nu_errors['ols']['a'].append(abs((model.B_ols[1] - a)/a))
        nu_errors['huber1']['a'].append(np.abs((model.B_huber_1[1] - a)/a))
        nu_errors['huber2']['a'].append(np.abs((model.B_huber_2[1] - a)/a))
        nu_errors['huber3']['a'].append(np.abs((model.B_huber_3[1] - a)/a))
        nu_errors['huber4']['a'].append(np.abs((model.B_huber_4[1] - a)/a))
        
        dof_estimates_tmp.append(model.nu)
    
    # compute average errors
    for name in errors:
        for coeff in errors[name]:
            errors[name][coeff].append(np.mean(nu_errors[name][coeff]))
            
    # compute average nu estimates
    dof_estimates.append(np.mean(dof_estimates_tmp))
```

```python
fig, axes = plt.subplots(2,1)

for name in errors:
    axes[0].plot(nus, errors[name]['a'], 'o-', label=name)
    axes[1].plot(nus, errors[name]['b'], 'o-', label=name)

for ax in axes:
    ax.plot(nus, dof_estimates, '+--')
    ax.legend()
    ax.set_ylim([0,2])
    
axes[0].set_title('Estimating Intercept')
axes[0].set_xlabel('Degrees of freedom')
axes[0].set_ylabel('Average error (%)')
axes[1].set_title('Estimating Slope')
axes[1].set_xlabel('Degrees of freedom')
axes[1].set_ylabel('Average error (%)')
    
plt.rcParams['figure.figsize'] = (15,20)
plt.show()
```
